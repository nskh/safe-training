```python
def gradient_descent(
    func, gradient, start, learn_rate, n_iter=1000, tolerance=1e-06, plot=True, figname=None, xlims=[], ylims=[], num=50,
):
    plt.figure()
    if plot:
        if xlims:
            xs = np.linspace(xlims[0], xlims[1], num)
        else:
            xs = np.linspace(-10, 10, num)
        plt.plot(xs, func(xs))
    vector = start
    estimates = np.array([start, func(start)])
    for _ in range(n_iter):
        # TODO: add noise to gradient? simulate stochasticity
        diff = -learn_rate * gradient(vector)
        if np.all(np.abs(diff) <= tolerance):
            break
        vector += diff
        estimates = np.vstack([estimates, np.array([vector, func(vector)])])
    plt.plot(estimates[:, 0], estimates[:, 1], 'g-o')
    plt.xlabel("$x$")
    plt.ylabel("$f(x)$")
    if xlims:
        plt.xlim(xlims)
    if ylims:
        plt.ylim(ylims)
    if figname is not None:
        assert type(figname) == str, "figname input was not string"
        if figname[-4:] == ".pdf":
            plt.savefig(figname)
        else:
            plt.savefig(figname + ".pdf")
    plt.show()
    return vector
```

```python
opt = gradient_descent(
    func = lambda v: v ** 2, gradient=lambda v: 2 * v, start=10.0, learn_rate=0.2, figname = "standard-gd-example.pdf"
)
print(f"Gradient descent returned (rounded) answer: {'{:.3f}'.format(opt)} and exact answer {'{:.3g}'.format(opt)}")
```

```python
def noisy_gradient_descent(
    func, gradient, start, learn_rate, noise_std=0.1, n_iter=1000, tolerance=1e-06, plot=True, figname=None,
    xlims=[], ylims=[], num=50
):
    plt.figure()
    if plot:
        if xlims:
            xs = np.linspace(xlims[0], xlims[1], num)
        else:
            xs = np.linspace(-10, 10, num)
        plt.plot(xs, func(xs))
    vector = start
    estimates = np.array([start, func(start)])
    for _ in range(n_iter):
        grad_result = gradient(vector)
        if type(grad_result) is float:
            diff = -learn_rate * grad_result * (1 + np.random.randn() * noise_std)
        else:
            diff = -learn_rate * grad_result * (1 + np.random.randn(grad_result.shape) * noise_std)
        if np.all(np.abs(diff) <= tolerance):
            break
        vector += diff
        estimates = np.vstack([estimates, np.array([vector, func(vector)])])
    plt.plot(estimates[:, 0], estimates[:, 1], 'g-o')
    plt.xlabel("$x$")
    plt.ylabel("$f(x)$")
    if figname is not None:
        assert type(figname) == str, "figname input was not string"
        if figname[-4:] == ".pdf":
            plt.savefig(figname)
        else:
            plt.savefig(figname + ".pdf")
    plt.show()
    return vector
```

```python
def constrained_gradient_descent_redraw(
    func, gradient, start, learn_rate, noise_std=0.1, n_iter=1000, tolerance=1e-06, plot=True, figname=None,
    xlims=[], ylims=[], num=50, test=lambda x: True,
):
    MAX_REDRAW = 10
    plt.figure()
    if plot:
        if xlims:
            xs = np.linspace(xlims[0], xlims[1], num)
        else:
            xs = np.linspace(-10, 10, num)
        plt.plot(xs, func(xs))
    vector = start
    estimates = np.array([start, func(start)])
    for _ in range(n_iter):
        counter = MAX_REDRAW
        while counter > 0:
            counter -= 1
            if counter == 0:
                print(f"had to stop training because safety constraint failed {MAX_REDRAW} times")
            grad_result = gradient(vector)
            if type(grad_result) is float:
                diff = -learn_rate * grad_result * (1 + np.random.randn() * noise_std)
            else:
                diff = -learn_rate * grad_result * (1 + np.random.randn(grad_result.shape) * noise_std)
            if np.all(np.abs(diff) <= tolerance):
                break
            if test(vector + diff):
                vector += diff
                estimates = np.vstack([estimates, np.array([vector, func(vector)])])
                break
    plt.plot(estimates[:, 0], estimates[:, 1], 'g-o')
    plt.xlabel("$x$")
    plt.ylabel("$f(x)$")
    if figname is not None:
        assert type(figname) == str, "figname input was not string"
        if figname[-4:] == ".pdf":
            plt.savefig(figname)
        else:
            plt.savefig(figname + ".pdf")
    plt.show()
    return vector
```

```python
def gradient_descent(
    cost, gradient, x, y, start, learn_rate=0.1, n_iter=50, tolerance=1e-06,
    dtype="float64"
):
    # Checking if the gradient is callable
    if not callable(gradient):
        raise TypeError("'gradient' must be callable")

    # Setting up the data type for NumPy arrays
    dtype_ = np.dtype(dtype)

    # Converting x and y to NumPy arrays
    x, y = np.array(x, dtype=dtype_), np.array(y, dtype=dtype_)
    if x.shape[0] != y.shape[0]:
        raise ValueError("'x' and 'y' lengths do not match")

    # Initializing the values of the variables
    vector = np.array(start, dtype=dtype_)

    # Setting up and checking the learning rate
    learn_rate = np.array(learn_rate, dtype=dtype_)
    if np.any(learn_rate <= 0):
        raise ValueError("'learn_rate' must be greater than zero")

    # Setting up and checking the maximal number of iterations
    n_iter = int(n_iter)
    if n_iter <= 0:
        raise ValueError("'n_iter' must be greater than zero")

    # Setting up and checking the tolerance
    tolerance = np.array(tolerance, dtype=dtype_)
    if np.any(tolerance <= 0):
        raise ValueError("'tolerance' must be greater than zero")
        
    estimates = np.array(start, dtype=dtype_)
    costs = np.array(cost(x, y, start), dtype=dtype_)

    # Performing the gradient descent loop
    for _ in range(n_iter):
        # Recalculating the difference
        diff = -learn_rate * np.array(gradient(x, y, vector), dtype_)

        # Checking if the absolute difference is small enough
        if np.all(np.abs(diff) <= tolerance):
            break

        # Updating the values of the variables
        vector += diff
        estimates = np.vstack([estimates, vector])
        costs = np.vstack([costs, cost(x, y, vector)])
    
    return vector if vector.shape else vector.item(), estimates, costs
```

```python
def cost(x, y, b):
    return np.mean((y - b[0] - b[1] * x) ** 2) / 2

def cost2d(x, y, b0, b1):
    return cost(x, y, [b0, b1])

def ssr_gradient(x, y, b):
    res = b[0] + b[1] * x - y
    return res.mean(), (res * x).mean()  # .mean() is a method of np.ndarray
```

```python
def gradient_descent_projection(
    cost, gradient, x, y, start, learn_rate=0.1, n_iter=50, tolerance=1e-06,
    dtype="float64", region=None,
):
    # Checking if the gradient is callable
    if not callable(gradient):
        raise TypeError("'gradient' must be callable")

    # Setting up the data type for NumPy arrays
    dtype_ = np.dtype(dtype)

    # Converting x and y to NumPy arrays
    x, y = np.array(x, dtype=dtype_), np.array(y, dtype=dtype_)
    if x.shape[0] != y.shape[0]:
        raise ValueError("'x' and 'y' lengths do not match")

    # Initializing the values of the variables
    vector = np.array(start, dtype=dtype_)

    # Setting up and checking the learning rate
    learn_rate = np.array(learn_rate, dtype=dtype_)
    if np.any(learn_rate <= 0):
        raise ValueError("'learn_rate' must be greater than zero")

    # Setting up and checking the maximal number of iterations
    n_iter = int(n_iter)
    if n_iter <= 0:
        raise ValueError("'n_iter' must be greater than zero")

    # Setting up and checking the tolerance
    tolerance = np.array(tolerance, dtype=dtype_)
    if np.any(tolerance <= 0):
        raise ValueError("'tolerance' must be greater than zero")
        
    estimates = np.array(start, dtype=dtype_)
    costs = np.array(cost(x, y, start), dtype=dtype_)

    # Performing the gradient descent loop
    for _ in range(n_iter):
        # Recalculating the difference
        diff = -learn_rate * np.array(gradient(x, y, vector), dtype_)

        # Checking if the absolute difference is small enough
        if np.all(np.abs(diff) <= tolerance):
            break

        # Updating the values of the variables
        vector += diff
        
        # TODO fix hardcoding for dimension
        if region:
            if vector[0] < region["x"][0]:
                print("projecting")
                vector[0] = region["x"][0]
            elif vector[0] > region["x"][1]:
                print("projecting")
                vector[0] = region["x"][1]

            if vector[1] < region["y"][0]:
                print("projecting")
                vector[1] = region["y"][0]
            elif vector[1] > region["y"][1]:
                print("projecting")
                vector[1] = region["y"][1]
            
        estimates = np.vstack([estimates, vector])
        costs = np.vstack([costs, cost(x, y, vector)])
    
    return vector if vector.shape else vector.item(), estimates, costs
```

```python
start = [0.5, 0.5]
params, estimates, costs = gradient_descent_projection(
    cost, ssr_gradient, x, y, start=start, learn_rate=0.0015,
    n_iter=100000, region={"x": [4, 6], "y": [0.4, 0.6]}
)
params
```

```python
def gridcost(xs, ys):
    res = np.zeros((ys.shape[0], xs.shape[0]))
    for i, xp in enumerate(xs):
        for j, yp in enumerate(ys):
            res[j, i] = cost2d(x, y, xp, yp)
    return res

def gridgrad(xs, ys):
    res = np.zeros((ys.shape[0], xs.shape[0]))
    for i, xp in enumerate(xs):
        for j, yp in enumerate(ys):
            res[j, i] = np.linalg.norm(ssr_gradient(x, y, (xp, yp)))
    return res
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

# xmin = estimates[:, 0].min()
# xmax = estimates[:, 0].max()
# ymin = estimates[:, 1].min()
# ymax = estimates[:, 1].max()

xmin, xmax = -10, 20
ymin, ymax = -1, 2
numpoints = 100

X = np.linspace(xmin, xmax, numpoints)
Y = np.linspace(ymin, ymax, numpoints)
costs = gridcost(X, Y)

# Make data.
Xmesh, Ymesh = np.meshgrid(X, Y)
surf = ax.plot_surface(Xmesh, Ymesh, costs, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
plt.xlabel('x')
plt.ylabel('y')
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

# xmin = estimates[:, 0].min()
# xmax = estimates[:, 0].max()
# ymin = estimates[:, 1].min()
# ymax = estimates[:, 1].max()

xmin, xmax = -10, 20
ymin, ymax = -1, 2
numpoints = 100

X = np.linspace(xmin, xmax, numpoints)
Y = np.linspace(ymin, ymax, numpoints)
costs = gridcost(X, Y)
grads = gridgrad(X, Y)

minn, maxx = grads.min(), grads.max()
norm = matplotlib.colors.Normalize(minn, maxx)
m = plt.cm.ScalarMappable(norm=norm, cmap='jet')
m.set_array([])
fcolors = m.to_rgba(grads)

# Make data.
Xmesh, Ymesh = np.meshgrid(X, Y)
ax.plot_surface(Xmesh, Ymesh, costs, rstride=1, cstride=1, facecolors=fcolors, vmin=minn, vmax=maxx, shade=False)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('cost')

plt.title("Reward Surface, colored by gradient")
plt.tight_layout()
plt.show()
```
